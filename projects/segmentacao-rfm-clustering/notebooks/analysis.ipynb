## Objetivo
Segmentar base de clientes para definir estrategias comerciais por perfil de valor.

## Perguntas de negocio
- Qual k gera melhor separacao?
- Quais personas concentram receita?
- Como transformar clusters em campanhas objetivas?


```python
from pathlib import Path
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

project_root = Path('..')
data_path = project_root / 'data' / 'rfm_clusters.csv'
metrics_path = project_root / 'models' / 'metrics.json'

pd.set_option('display.max_columns', 200)
df = pd.read_csv(data_path)
metrics = json.loads(metrics_path.read_text(encoding='utf-8')) if metrics_path.exists() else {}

print('shape:', df.shape)
print('metrics keys:', list(metrics.keys()))
df.head(5)
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing = missing[missing > 0]
print('missing columns:', len(missing))
if len(missing) > 0:
    display(missing)

if 'cluster' in df.columns:
    print('target distribution:')
    print(df['cluster'].value_counts(normalize=True).round(4))
```

```python
cluster_stats = df.groupby('persona').agg(
    customers=('customer_id', 'count'),
    recency_avg=('recency_days', 'mean'),
    frequency_avg=('frequency', 'mean'),
    monetary_avg=('monetary', 'mean'),
).sort_values('customers', ascending=False)
cluster_stats
```

```python
plt.figure(figsize=(8, 4))
cluster_stats['customers'].plot(kind='bar')
plt.title('Clientes por persona')
plt.grid(axis='y', alpha=0.2)
plt.tight_layout()

plt.figure(figsize=(6, 4))
plt.scatter(df['frequency'], df['monetary'], c=df['cluster'], s=8, alpha=0.35)
plt.title('Frequency x Monetary por cluster')
plt.grid(alpha=0.2)
plt.tight_layout()
```
